```python
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

filepath = '~/Final/Tara_Oceans_Surface_Water_Measurements.csv'
ocean_data = pd.read_csv(filepath, low_memory=False) # low memory=false was required to load
#display(ocean_data)

# Temperature
fig1 = plt.figure(figsize=(10, 5))
ax1 = plt.axes(projection=ccrs.PlateCarree())
ax1.add_feature(cfeature.LAND, color='wheat') # added land and color for land (temp fig)
ax1.add_feature(cfeature.OCEAN, color='steelblue') # added ocean and color for ocean (temp fig)
ax1.coastlines() # coastlines for temp fig

temp = ax1.scatter(ocean_data['lon'], ocean_data['lat'], c=ocean_data['temperature'], cmap='inferno')
plt.colorbar(temp, label='Temperature (C)')
plt.title('Distribution of Temperature in Ocean', fontsize='14')
g1 = ax1.gridlines(crs=ccrs.PlateCarree(), color='white', linestyle='--', alpha=0.5)
g1.bottom_labels=True # longitude
g1.left_labels=True # latitude

# Chlorophyll A
fig2 = plt.figure(figsize=(10, 5))
ax2 = plt.axes(projection=ccrs.PlateCarree())
ax2.add_feature(cfeature.LAND, color='wheat') # added land and color for land (chl fig)
ax2.add_feature(cfeature.OCEAN, color='steelblue') # added ocean and color for ocean (chl fig)
ax2.coastlines() # coastlines for chl fig

chl = ax2.scatter(ocean_data['lon'], ocean_data['lat'], c=ocean_data['chl_a'], cmap='viridis_r') #had to reverse gradient because it was difficult to see
plt.colorbar(chl, label='Chlorophyll A (mg/m^3)')
plt.title('Distribution of Chlorophyll A in Ocean', fontsize='14')
g2 = ax2.gridlines(crs=ccrs.PlateCarree(), color='white', linestyle='--', alpha=0.5)
g2.bottom_labels=True # longitude
g2.left_labels=True # latitude

#Originally two seperate graphs so I just combined the two to show the trend
```

```python
import pandas as pd
import matplotlib.pyplot as plt

filepath = '~/Final/Tara_Oceans_Surface_Water_Measurements.csv'
ocean_data = pd.read_csv(filepath, low_memory=False) # low memory=false was required to load
#display(ocean_data)

fig, ax1 = plt.subplots(figsize=(15, 5))
plt.title('Temperature and Chlorophyll A by Latitude', fontsize=18)

# Temperature
ax1.scatter(ocean_data['lat'], ocean_data['temperature'], c='firebrick', alpha=0.5, marker='.')
ax1.set_xlabel('Latitude', fontsize=14)
ax1.set_ylabel('Temperature (C)', color='firebrick', fontsize=14)
ax1.tick_params(axis='y', labelcolor='firebrick')
ax1.grid(linestyle='--', color='white', alpha=0.5)
ax1.set_facecolor('skyblue')
ax1.legend(['Temperature'], loc='upper left')

# Chlorophyll A
ax2 = ax1.twinx()
ax2.scatter(ocean_data['lat'], ocean_data['chl_a'], c='seagreen', alpha=0.5, marker='.')
ax2.set_ylabel('Chlorophyll a (mg/m^3)', color='seagreen', fontsize=14)
ax2.tick_params(axis='y', labelcolor='seagreen')
ax2.legend(['Chlorophyll A'], loc='upper right')

plt.show()
```
